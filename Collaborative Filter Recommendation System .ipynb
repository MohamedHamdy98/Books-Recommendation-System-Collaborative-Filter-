```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data_books = pd.read_csv(r'I:\Python\dataSet\ml\book recommendation\Books.csv')
data_rating =  pd.read_csv(r'I:\Python\dataSet\ml\book recommendation\Ratings.csv')
```

```python
main_data = data_books.merge(data_rating, on='ISBN', how='inner')
```

```python
data = main_data.copy()
data.head(5)
```

```python
data.isna().sum()
```

```python
data.rename(columns={'Book-Title':'title', 'User-ID':'user_id', 'Book-Rating':'rating'}, inplace=True)
```

```python
data.head(2)
```

```python
data.shape
```

```python
data['rating'].unique()
```

```python
data = data.loc[data['rating'] > 4]
```

```python
data.shape
```

```python
data = data.iloc[:10000, :]
data.shape
```

```python
utility_matrix = data.pivot_table(values='rating', index='user_id', columns='title') 
```

```python
utility_matrix
```

```python
utility_matrix.fillna(0, inplace=True)
```

```python
utility_matrix
```

```python
cosine_sim_user = cosine_similarity(utility_matrix)
```

```python
user_id = data['user_id']
indices = pd.Series(user_id.index, index=data['user_id'])
```

```python
cosine_dataframe_user = pd.DataFrame(cosine_sim_user, index=utility_matrix.index, columns=utility_matrix.index)
cosine_dataframe_user
```

```python
def get_recommendations(user_id,cosine_sim=cosine_sim_user):
    idx = indices[user_id]
    sim_scores =list(enumerate(cosine_sim_user[idx]))
    sim_scores =sorted(sim_scores, key=lambda x : x[1], reverse=True)
    sim_scores =sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    result = data[['user_id','title','Book-Author','Publisher','rating']].iloc[movie_indices]
    return result
```
